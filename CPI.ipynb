CPI dataは統計局のデータベースから取得。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df_CPI = pd.read_excel('data/CPI Year AVG.xlsx', sheet_name = 'bn01-2 (2)', header=13, index_col=8)
```

```python
df_CPI_all = df_CPI.iloc[:, 11]
df_CPI_all
```

```python
df_CPI_all_new = df_CPI_all[11:]
df_CPI_all_new.index = df_CPI_all_new.index.str[:4].astype(int)
df_CPI_all_new
```

```python
df_CPI_all_new.to_excel('result/df_CPI_all.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(df_CPI_all_new)
ax.set_ylabel("％")
ax.set_xlabel('Year')
ax.axhline(y=3, color='red', linestyle='--')
ax.axhline(y=0, color='green', linestyle='--')
plt.title('CPI Rise in Japan (％）')

plt.grid()
plt.savefig("CPI.png")   
```
