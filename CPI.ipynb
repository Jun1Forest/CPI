日経平均株価を　St. Louis中銀から取る。それを処理。古いデータはYahoo等を参照。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df_Nikkei = pd.read_csv('data/NIKKEI225.csv')
df_Nikkei.columns =["date", "NIKKEI225"]
df_Nikkei['date'] = pd.to_datetime(df_Nikkei['date'].astype(str))
df_Nikkei = df_Nikkei.set_index('date')
df_Nikkei
```

```python
df_Nikkei_cleaned= df_Nikkei[df_Nikkei['NIKKEI225']!='.']
#df_Nikkei_cleaned = df_Nikkei_cleaned.astype(float)
df_Nikkei_cleaned.info()
```

```python

df_Nikkei_cleaned['NIKKEI225']= df_Nikkei_cleaned['NIKKEI225'].astype(float)
```

```python
df_Nikkei_cleaned.to_excel('result/df_Nikkei225_cleaned.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(df_Nikkei_cleaned['NIKKEI225'])
ax.set_ylabel("NIKKEI225")
ax.set_xlabel('Year')
plt.title('NIKKEI225')
plt.grid()
plt.savefig("Nikkei225.png")   
```
